## General Energy Balance Module

### System

Consider a process in which there are two input streams and two output streams, each at a different temperature.

#### Mass Balance Data
The inlet and outlet streams correspond to solutions with the following composition, temperature and flow rate:
* Stream 1: A (10% molar) + D, $T_1 = 0 ºC$, $F_1 = 100 mol/h$
* Stream 2: B (15% molar) + D, $T_2 = 50 ºC$, $F_2 = 100 mol/h$
* Stream 3: D pure, $T_3 = 100 ºC$, $F_3 = 50 mol/h$
* Stream 4: (A), B, (C), D, $T_4 = 75 ºC$, $F_4$ is calculated by balance
The following reaction can occur:
$$ A + B \rightarrow 2 C$$
With conversion = 0%, 50% or 100% (so that A and/or C may be present in stream 4).

#### Thermodynamic data
$$ \Delta H_f^0(A) = -100 kJ/mol$$
$$ \Delta H_f^0(B) = -200 kJ/mol$$
$$ \Delta H_f^0(C) = -1000 kJ/mol$$
$$ \Delta H_f^0(D) = -500 kJ/mol$$
$$ Cp(A) = 5 kJ/mol ºC$$
$$ Cp(B) = 2 kJ/mol ºC$$
$$ Cp(C) = 1 kJ/mol ºC$$
$$ Cp(D) = 1 kJ/mol ºC$$
Then the standard heat of reaction is:
$$ \Delta H_R^0 = -1700 kJ/mol$$

#### Problem Description
It is desired to calculate the heat $Q$ exchanged with the outside in the process, considering three assumptions:
* Process without chemical reaction (conversion = 0)
* Process with chemical reaction (conversion = 50%)
* Process with chemical reaction (conversion = 100%)

The energy balance for the system is expressed as:
$$ \dot Q = H_3 + H_4 - H_1 - H_2$$

Where the enthalpy of each stream ($H_i$) must be calculated by selecting one of the following expressions:
* Expression 1:
$$H_i = \sum_{component s} F_{i,s} \cdot Cp(s) \cdot (T_i - T_{ref})$$
* Expression 2 (method of heat of formation):
$$H_i = \sum_{component s} F_{i,s} \cdot (\Delta H_f^0(s) + Cp(s) \cdot (T_i - T_{ref}))$$

These expressions can be compared with the expression of the heat of reaction method:
$$\dot Q(heat) = \xi \cdot \Delta H_{Rx}^0 + \dot Q(expression 1)$$
